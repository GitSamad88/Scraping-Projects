```python
j=0
while True:
    try:
        result=requests.get(f'https://www.moteur.ma/fr/voiture/achat-voiture-occasion/{j}')
        # 3s: save page content/markup
        src= result.content

        # 4s : create soup object to parse content
        soup=BeautifulSoup(src,"lxml")
        page_number = int (soup.find ('span', {"class": "pull-right"}).findAll ('b')[1].text)
        car_3info = soup.find_all ('ul', {'style': 'font-size:11px;'})
        #5s: get car link
        car_link = soup.find_all ('div', {"class": "content-inner"})
        for i in range (len (car_link)):
            cars_links.append (car_link[i].find ('a', {"class": "slide"}))
        for i in range (len (cars_links)):
            Car_link.append (cars_links[i].attrs['href'])
        pd.DataFrame({"Links":Car_link}).to_csv("Cars_Links1.csv")

        if (j>160):
                print("end page!")
                break
        print(j)
        print(result.url)
        j=j+15


        print("page:",j//15,"page switched")


    except:
        print("erorr")

Car_link=list(dict.fromkeys(Car_link))
print(Car_link)
print(len(Car_link))
```

### extract each car data using its link

```python
for k,link in enumerate(list(Car_link)):
   
    cleaned_options=[]
    founded_options_list=[]
    try:

        result1 = requests.get (link)
        src1 = result1.content
        soup1 = BeautifulSoup (src1, "lxml")
        print(k,link)
        #print(type(soup1))
        """cars data and model"""
        mark_model = list (soup1.find ('div', {"class": "col-md-12 col-sm-12 col-xs-12 text-center ads-detail"}).find ('h1'))
        #print(soup1.find ('div', {"class": "col-md-12 col-sm-12 col-xs-12 text-center ads-detail"}))
        #print(soup1.find ('div', {"class": "col-md-12 col-sm-12 col-xs-12 text-center ads-detail"}).find('h1'))

        for i in mark_model[1:2]:
            Mark_Model.append (i.text.strip ( ))
            '''Get description'''
        #descriptions.append (soup1.find('div', {"class": "col-md-12", "style": "margin-top:1px;margin-bottom:19px;"}))

        '''get the price'''

        prices=soup1.find("div",{"class":"color_primary text_bold price-block"})
        if (prices!=None):Prices.append(prices.text.strip())
        else:Prices.append(np.nan)

        '''get car location'''

        Location.append(soup1.find("i",{"class":"icon icon-normal-pointer"}).next.strip())
        '''get seller name or nick_name'''
        seller_name.append (soup1.find ('a', {'style': 'font-size: 15px;'}).text.strip ( ))

        '''scrap first use data'''

        first_hand = soup1.find ('span', {'class': 'col-md-6 col-xs-6'},string="Première main") or soup1.find ('span',{'class': 'col-md-6 col-xs-6'}, string=re.compile ("Importé neuf"))

        if (first_hand != None):
            Première_main.append(1) #(first_hand.find_next_sibling ( ).text.strip ( ))
        else:
            Première_main.append (0)

        '''scrap kilometrage, carosserie, speed..., look to details list'''

        for i in range (len (details)):
            Detail = soup1.find ('span', {'class': 'col-md-6 col-xs-6'}, string=str (details[i]))
            if (Detail != None): details_list[i].append (Detail.find_next_sibling ( ).text.strip ( ))
            else: details_list[i].append (np.nan)

        '''extract cars featurs:Abs,Airbags...'''

        #check if the features exist and append 0(not exist) or 1(exist) to a list
        options=list(soup1.find_all('div', {"class":"col-md-6 option_ad"}))
        for i in options:
            founded_options_list.append(i.text.strip())
        for option in founded_options_list:
            clean_option=option.replace("✔","").replace(" ","_").replace("__","").replace("/","").replace("__","_")
            cleaned_options.append(clean_option) 
        print("cleaned_options: ",len(cleaned_options))

        for option in options_list:
            if option in cleaned_options:
                final_list[option].append(1)
            else :
                final_list[option].append(0)
         # add link:
        Link.append(link)
    except:
        print("erorr!")
        error_link.append(link)
        continue


    '''get seller number phone'''
    '''
    try:

        driver.get(link)
        print(k,":",link)
        sleep(5)
        driver.find_element_by_xpath ('//*[@id="overview"]/div[2]/ul/li[2]/a/div').click()
        sleep(5)
        phone = driver.find_element_by_class_name ('adcall').text
        print(phone)
        sleep(3)
        N_phone.append (phone)
    except:
        N_phone.append ("No Number")
        continue
    '''
```
