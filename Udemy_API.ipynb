```python
import requests
import json

# Set your API credentials
client_id="mVkPpIrcBjUZuvGfLY07Fof0yb0JjKiPJGmDkK4Q"
client_secret="VvJY0AuoNvTj15EeWdUeZ4MsnoXcUAh2zp6VBCjOrO82cfXtqEWpidUWMsXMymaemKkdihx19MaivCXJh0nYyDpseGl1umQT9lTxCEszF1CfqVVfcgxYI2C6KE9tfpGv"

# Get an access token
auth_url = "https://www.udemy.com/api-2.0/"
auth_data = {
    "client_id": client_id,
    "client_secret": client_secret,
    "grant_type": "client_credentials"
}
auth_response = requests.post(auth_url, data=auth_data)
#access_token = auth_response.json()["access_token"]

# Define the API endpoint and parameters
endpoint = "https://www.udemy.com/api-2.0/courses/12345/"
params = {
    "fields": "title,description,url"
}

# Make a request to the API
headers = {"Authorization": f"Bearer {access_token}"}
response = requests.get(endpoint, headers=headers, params=params)
print(response)
# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()
    # Print the title, description, and URL of the course
    print("Title:", data["title"])
    print("Description:", data["description"])
    print("URL:", data["url"])
else:
    print("Error:", response.status_code)
```
